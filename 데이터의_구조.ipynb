범주형 데이터 : 숫자값으로 의미를 가지지 않는 데이터
ex) 0과 1, 1,2,3,4,5 점수 매기기


```python
df1['카테고리'].value_counts()
```

```python
df1['카테고리'].describe()
```

```python
df1.describe().columns # 숫자 데이터의 column 이름만 가져올 수 있음
```

```python
df1.describe().columns.tolist() # 수치형 데이터의 리스트
```

```python
df1.columns.tolist() # 전체 데이터의 리스트
```

```python
total_list=df1.columns.tolist()
numeric_list=df1.describe().columns.tolist()
categorical_list=list(set(total_list)-set(numeric_list)) # 전체 데이터 - 수치 데이터
categorical_list # 범주형 데이터
```
