# 데이터 정렬

```python
# sort_values 을 이용해 정렬을 하고, by=' ' 정렬하고 싶은 컬럼을 기준으로 정렬하게 됨, 데이터를 내림 차순으로 하고 싶다
df1.sort_values(by='구매금액', ascending=False)
```

```python
df1.sort_values(by='구매수량', ascending=False)
```

```python
# 정렬 기준이, 구매수량과 연령인데, 여기서 구매수량을 우선순위를 두고 정렬한 다음 연령을 다음 우선순위로 잡아
# 정렬한다.
df1.sort_values(by=['구매수량','연령'], ascending=False)
```

```python
df1.sort_values(by=['구매수량','연령'], ascending=[False,True]).iloc[:10]
```

# 연습

* 30대 여성 추출 => 구매 금액 높은 순서로 정렬
* 파일로 저장

```python
cond1 = (df1['연령대']=='30대이하')
cond2 = (df1['성별']=='여')
df1.loc[cond1&cond2].sort_values(by='구매금액', ascending=False)
df1.shape
```
