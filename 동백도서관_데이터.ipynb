```python
# 1. 한글 폰트 설치
!apt-get -qq install -y fonts-nanum

# 2. 폰트를 matplotlib에 적용
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 설치한 폰트 사용
font_path = '/usr/share/fonts/truetype/nanum/NanumGothic.ttf'
fontprop = fm.FontProperties(fname=font_path, size=10)
plt.rc('font', family='NanumGothic')

# 3. 계절별 데이터 분석 및 시각화
import pandas as pd

# 파일 경로
file_path_autumn = '/content/동백도서관 장서 대출목록 (2023년 09월).xlsx'
file_path_winter = '/content/동백도서관 장서 대출목록 (2023년 12월).xlsx'
file_path_spring = '/content/동백도서관 장서 대출목록 (2024년 03월).xlsx'
file_path_summer = '/content/동백도서관 장서 대출목록 (2024년 06월).xlsx'
# 데이터 로드
df_autumn = pd.read_excel(file_path_autumn)
df_winter = pd.read_excel(file_path_winter)
df_spring = pd.read_excel(file_path_spring)
df_summer = pd.read_excel(file_path_summer)

# 계절별 데이터 분석 함수
def analyze_season_data(df, season_name):
    # 도서명과 주제분류번호를 기준으로 대출건수를 집계
    popular_books = df.groupby('도서명')['대출건수'].sum().sort_values(ascending=False).head(10)
    popular_subjects = df.groupby('주제분류번호')['대출건수'].sum().sort_values(ascending=False).head(10)

    # 시각화
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    popular_books.plot(kind='bar', color='skyblue')
    plt.title(f'{season_name} 인기 도서 Top 10', fontproperties=fontprop)
    plt.xlabel('도서명', fontproperties=fontprop)
    plt.ylabel('대출 건수', fontproperties=fontprop)
    plt.xticks(rotation=45, ha='right', fontproperties=fontprop)

    plt.subplot(1, 2, 2)
    popular_subjects.plot(kind='bar', color='lightgreen')
    plt.title(f'{season_name} 인기 주제 Top 10', fontproperties=fontprop)
    plt.xlabel('주제', fontproperties=fontprop)
    plt.ylabel('대출 건수', fontproperties=fontprop)
    plt.xticks(rotation=45, ha='right', fontproperties=fontprop)

    plt.tight_layout()
    plt.show()

    return popular_books, popular_subjects

# 분석 및 시각화
autumn_books, autumn_subjects = analyze_season_data(df_autumn, '가을')
winter_books, winter_subjects = analyze_season_data(df_winter, '겨울')
spring_books, spring_subjects = analyze_season_data(df_spring, '봄')
summer_books, summer_subjects = analyze_season_data(df_summer, '여름')
```
